```python
import nltk
```

```python
sentence = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"),("dog", "NN"), ("barked", "VBD"), ("at", "IN"),  ("the", "DT"), ("cat", "NN")]
grammar = "NP: {<DT>?<JJ>*<NN>}"
grammar1 = "VP: {<PRP>+<VB>+<JJ>?}"
```

```python
cp = nltk.RegexpParser(grammar)
cp1 = nltk.RegexpParser(grammar1)
```

```python
result = cp.parse(sentence)
```

```python
print(result)
```
